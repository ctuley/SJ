```python
#Recoveries
df["type/name"].replace({"Ball Recovery": "Recovery"}, inplace=True)
df6=df[(df['type/name']=="Recovery")]
df6 = df6[df6['location/0']<102]
df6 = df6[df6['location/0']>18]
df6 = df6[df6['location/1']>18]
df6 = df6[df6['location/1']<62]
```

```python
#Shots
df7=df[(df['type/name']=="Shot")]
```

```python
data= {'Passes Attempted':[len(df[df['type/name'] == 'Pass'])],
       'Carries':[len(df[df['type/name'] == 'Carry'])],
       'Clearances':[len(df[df['type/name'] == 'Clearance'])],
       'Incomplete Passes':[len(df[df['pass/outcome/name'] == 'Incomplete'])],
       'Passes from Def3rd to Mid3rd':[len(df1[df1['type/name'] == 'Pass'])],
       'Incomplete Passes from Def3rd to Mid3rd':[len(df1[df1['pass/outcome/name'] == 'Incomplete'])],
       'Passes from Def3rd to Fin3rd Attempted':[len(df2[df2['type/name'] == 'Pass'])],
       'Incomplete Passes from Def3rd to Fin3rd':[len(df2[df2['pass/outcome/name'] == 'Incomplete'])],
       'Passes from Mid3rd to Fin3rd':[len(df3[df3['type/name'] == 'Pass'])],
       'Incomplete Passes from Mid3rd to Fin3rd':[len(df3[df3['pass/outcome/name'] == 'Incomplete'])],
       'Passes Under Pressure Attempted':[len(df4[df4['type/name'] == 'Pass'])],
       'Incomplete Passes Under Pressure':[len(df4[df4['pass/outcome/name'] == 'Incomplete'])],
       'Forward Passes Attempted':[len(df5[df5['type/name'] == 'Pass'])],
       'Incomplete Forward Passes':[len(df5[df5['pass/outcome/name'] == 'Incomplete'])],
       'Ball Recoveries':[len(df[df['type/name'] == 'Recovery'])],
       'Ball Recoveries Between Boxes':[len(df6[df6['type/name'] == 'Recovery'])],
       'xG':[df['shot/statsbomb_xg'].sum()],
       'Shots Attempted':[len(df[df['type/name'] == 'Shot'])],
       'Shots Attempted From Open Play':[len(df7[df7['shot/type/name'] == 'Open Play'])],
       'Goals':[len(df7[df7['shot/outcome/name'] == 'Goal'])],
       'Shots Saved':[len(df7[df7['shot/outcome/name'] == 'Saved'])],
       'Penalty Goals':[len(df8[df8['shot/outcome/name'] == 'Goal'])],
       'Penalty xG':[df8['shot/statsbomb_xg'].sum()]
       }

Gabr = pd.DataFrame(data=data)
```
